# ML-03 — ML Task Framing

## Lane: Content Refresh Prioritization

This notebook maps the selected lane onto a concrete machine-learning task and connects the model output to a real content-team decision.

## 1. My lane as an ML task

My lane is **Content Refresh Prioritization**.

This is mainly a **ranking problem** because the final goal is to rank webpages from highest to lowest priority for manual review and possible content refresh.

A classification model can estimate whether a page is likely to be declining, but the useful business output is still a ranked list because a content team cannot review every page at once.

## 2. Target or proxy

My provisional target is whether a webpage has a declining search-performance trend.

The starter dataset contains a proxy column called `trend_direction`. I will create a binary target:

- `1` when `trend_direction` is `down`
- `0` otherwise

This is only a proxy for review priority. It does not prove that the content itself is outdated, and it does not guarantee that refreshing the page will improve its performance.

## 3. Success metric

My main success metric is **Precision@50**.

Precision@50 measures how many of the top 50 recommended pages are actually labelled as declining.

This metric matches the real action because a content team may only have time to review a small number of pages first. A useful system should place the most relevant pages near the top of the ranked list.

## 4. The unit of analysis, as a real dataframe

The unit of analysis is **one webpage**. Each row in the dataframe represents one anonymized webpage, while the columns contain page-level search and content signals.

```python
import pandas as pd

url = "https://raw.githubusercontent.com/Nav33dCodes/flyrank-ml-internship-naveed/main/data/raw/content_refresh_anonymized.csv"
df = pd.read_csv(url)

print("Dataset shape:", df.shape)
print("One row represents one webpage.")
display(df.head())
```

The dataframe shows page-level signals such as impressions, CTR, average position, content age, and trend direction. These signals can be used together to support refresh prioritization.

```python
# Sketch the target column
df["refresh_target"] = (df["trend_direction"] == "down").astype(int)

target_preview = df[
    [
        "trend_direction",
        "refresh_target",
        "impressions_90d",
        "ctr",
        "avg_position"
    ]
].head(10)

display(target_preview)

print("Target counts:")
print(df["refresh_target"].value_counts())
```

Here, `refresh_target = 1` means the page has a declining trend and may deserve review. `refresh_target = 0` means it is not labelled as declining.

The final output will be a **ranked list of webpages**. A content or SEO team will review the highest-ranked pages first and decide whether to update, improve, consolidate, investigate further, or leave each page unchanged. The system supports human decision-making and does not automatically change any content.

## 5. Why ML beats a fixed rule here

A fixed rule might say that every page with low CTR or old content should be refreshed. This can be misleading because search performance depends on several signals at the same time.

For example, a page may have low CTR because its average search position is poor, while another page may be old but still perform well. ML can combine impressions, CTR, average position, content age, and other observed signals instead of relying on one threshold.

This makes the task more suitable for ML or multi-signal analysis than for a single fixed rule. However, the model should still be compared with a transparent baseline rule so that any improvement is measured rather than assumed.
